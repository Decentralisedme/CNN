### Helper function

```python
# Helper_Function:create new tf variables in the given shape
# we need them to 
def new_weights(shape):
    return tf.Variable(tf.truncated_normal(shape, stddev=0.05))
```

```python
def new_biases(length):
    return tf.Variable(tf.constant(0.05, shape=[length]))
```

### Now we will have to implemtn the folowing steps:

### 1 New Convolutional Layer
###### 1A- created in the TF graph , but here are just formulas
###### 1B- assums imput is 4-dimention tensor
########## 1B1 Image number
########## 1B2 X axis each image
########## 1B3 Y axis each image
########## 1B4 Channel of each images
####### Channels can be colors channels or filter channel if input comes from anothe layer 
###### 1C- Output is a 4 dimentianal layer
########## 1C1 Image number
########## 1C2 X axis of each image, depends on the pooling
########## 1C3 Y axis of each image
########## 1C4 Channel oproduced by each conv filters

### 2 Flattening a layer
### 3 New Fully-Connected Layer


```python
# 1 New Convolutional Layer
def new_conv_layer(input, num_input_channels, filter_size, num_filters, use_pooling = True):
    shape = [filter_size, filter_size, num_input_channels, num_filters]
    weights = new_weights(shape = shape)
    biases = new_biases(length = num_filters)
    layer = tf.nn.conv2d(input = input, filter = weights, strides = [1, 1, 1, 1], padding = 'SAME')
    
    layer += biases
    
    if use_pooling:
        layer = tf.nn.max_pool(value = layer, ksize = [1,2,2,1], strides = [1,2,2,1], padding = 'SAME')
    layer = tf.nn.relu(layer)
    return layer, weights
    
```

```python
# 2 Flattening a layer >>> 1 produce 4-dimension tensor, we need a 2-Dimention layer to be an input for fully connected layer
def flatten_layer(layer):
    layer_shape = layer.get_shape()
    num_features = layer_shape[1:4].num_elements()
    layer_flat = tf.reshape(layer, [-1, num_features])
    return layer_flat, num_features
```

```python
# 3 New Fully-Connected Layer: we have output of previus layer as input of this ayer
def new_fc_layer(input, num_inputs, num_outputs, use_relu = True):
    weights = new_weights(shape = [num_inputs, num_outputs])
    biases = new_biases(length = num_outputs)
    layer = tf.matmul(input, weights) + biases
    
    if use_relu:
        layer = tf.nn.relu(layer)
    return layer
```

### 1- PlaceHolder Variuables

```python
# 1- PlaceHolder Variuable>> input images, is a tensor, multidimentional vectoror matrix with floats
x = tf.placeholder(tf.float32, shape=[None, img_size_flat], name='x')
```

### 5- Fully-Connected Layer 1

```python
# This is the fully connected layer, in which -input= previs flat layer and - nuber of node = fc_size
layer_fc1 = new_fc_layer(input=layer_flat,
                         num_inputs=num_features,
                         num_outputs=fc_size,
                         use_relu=True)
```

### 6- Fully-Connected Layer 2

```python
# add another Full Con Layer the output of which is a vector of 10 use to determone to which class the output belong
layer_fc2 = new_fc_layer(input=layer_fc1,
                         num_inputs=fc_size,
                         num_outputs=num_class,
                         use_relu=False)
```

```python
layer_fc2
```

### 7- Predicted Class

```python
## Full Con lay 2 estimates how likly each image belongs to each of the 10 different classes
## Cosenumber can be very large, we can normilse them, so value btwn 0 and 1, and their sum =1 >>
## We do this using softmax function 
y_pred = tf.nn.softmax(layer_fc2)
```

```python
y_pred 
```

```python
## We now use armax to pick the larger vaue of the ouput layer, which is the class with hihest prob,
## we use arg max
```

```python
y_pred_cls = tf.argmax(y_pred, dimension = 1)
```

```python
y_pred_cls
```

### 8- Cost-function to be optimized

```python
# we now want to know how the model is performming so we compare y_pred with y_true
# We use CROSS-ENTROPY = meansures how accurate the classificatons if the netfork are compared to the true calsses
# C_E is function thta is always >= 0: 0  means is correct, 
# so we want to minimise the funct (criss_entropy close to 0) by changeing the variables of Network Layer
# TN has cross entropy funct, hnd cose calc softmax, we must use layer fc 2 (y_pred calc already)
# this funct calc the ce for each input impage, but we want a single vsalue which will be the average accress
```

```python
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits = layer_fc2, labels=y_true)
```

```python
cross_entropy
```

```python
# we take the average
cost = tf.reduce_mean(cross_entropy)
```

```python
cost
```

```python
# we use Adam optimiser (Gradient Descent)
optimizer = tf.train.AdamOptimizer(learning_rate=1e-4).minimize(cost)
```

```python
total_iterations = 0

def optimize(num_iterations):
    # variables need to be updated globaly not localy
    global total_iterations
    
    #We print timing
    start_time = time.time()
    
    for i in range(total_iterations, total_iterations + num_iterations):
        # Get x_batch = batch of images and y_true_batch = batch of the true labels of these images
        x_batch, y_true_batch = data.train.next_batch(train_batch_size)
        
        # put the above in a dictfor lace holder variable in the graph
        feed_dict_train = {x: x_batch, y_true: y_true_batch}
        
        # Now we can run the session and otimiser
        # 1 - Run using this batch of the training data
        # 2 - TensorFlow assign:
        ###### variables from dictionary >> to >> placehokder
        # 3 - Run the Optimizer
        session.run(optimizer, feed_dict = feed_dict_train)
        
        # Print sattus
        if i % 100 == 0:
            # Accuracy
            acc = session.run(accuracy, feed_dict = feed_dict_train)
            
            #Message for printin
            msg = "Optimization Iteration: {0:>6}, Training Accuracy: {1:>6.1%}"
            
            # Print
            print(msg.format(i + 1, acc))
            
        # Update numbe of iteration
        total_iterations += num_iterations
        
        # Ending time
        end_time = time.time()
        
        # Time taken 
        time_dif = end_time - start_time
        
        #Print time usdage
        print("Time usage: " + str(timedelta(seconds=int(round(time_dif)))))
        
```

```python
def plot_confusion_matrix(cls_pred):
    # This is called from print_test_accuracy() below.

    # cls_pred is an array of the predicted class-number for
    # all images in the test-set.

    # Get the true classifications for the test-set.
    cls_true = data.test.cls
    
    # Get the confusion matrix using sklearn.
    cm = confusion_matrix(y_true=cls_true,
                          y_pred=cls_pred)

    # Print the confusion matrix as text.
    print(cm)

    # Plot the confusioncls_truetrix as an image.
    plt.matshow(cm)

    # Make various adjustments to the plot.
    plt.colorbar()
    tick_marks = np.arange(num_class)
    plt.xticks(tick_marks, range(num_class))
    plt.yticks(tick_marks, range(num_class))
    plt.xlabel('Predicted')
    plt.ylabel('True')

    # Ensure the plot is shown correctly with multiple plots
    # in a single Notebook cell.
    plt.show()
```

#### D - Show performace

```python
test_batch_size = 256


def print_test_accuracy(show_example_errors = False, show_confusion_matrix = False):
    # num of imag in test set
    num_test = len(data.test.images)
    
    # Array for pred classes: they will be calc in barches and put in this array
    cls_pred = np.zeros(shape=num_test, dtype= np.int_)
    
    # We calc the the predict class for the batches
    # we iterate through batches
#     y_true = tf.reshape(y_true[64,10])
    #start index. = i
    i = 0
    while i < num_test:
        j = min(i + test_batch_size, num_test)
        
        # Get img from test set btwn i and j
        images = data.test.images[i:j, :]
        
        # GEt associated labels
        labels = data.test.labels[i:j, :]
       
        # Create the dict with img and labe;s
        feed_dict = {x: images, y_true: labels}
        
        # Calc pred class
        cls_pred[i:j] = session.run(y_pred_cls, feed_dict = feed_dict)
        
        # set start index of next batch = end index of current batch
        i = j
        
    
    # Conv variable for the true class numb of the test set
    cls_true = data.test.cls
    
    #Creat bool to check if img correctly classified
    correct = (cls_true == cls_pred)
    
    #Calc the number of correctly classified img
    # we sum the 0s (Flase) and 1s(True)
    correct_sum = correct.sum()
    
    #Clac ACCURACY of correctly class imgs
    acc = float(correct_sum)/ num_test
    
    # Print the accuracy.
    msg = "Accuracy on Test-Set: {0:.1%} ({1} / {2})"
    print(msg.format(acc, correct_sum, num_test))

    # Plot some examples of mis-classifications, if desired.
    if show_example_errors:
        print("Example errors:")
        plot_example_errors(cls_pred=cls_pred, correct=correct)

    # Plot the confusion matrix, if desired.
    if show_confusion_matrix:
        print("Confusion Matrix:")
        plot_confusion_matrix(cls_pred=cls_pred)
```

### 1 - Weights

```python
# Plotting the weights
def plot_conv_weights(weights, input_channel = 0):
    # weights are Tesorflow ops for 4-dim variables
    # they ould be weights_conv1 or weights_conv2
    w = session.run(weights)
    
    # -1 Get min and max w values
    ###this is to give a scale to colors across the images
    w_min = np.min(w)
    w_max = np.max(w)
    
    # -2 Get num of filters to create a grid to plot
    num_filters = w.shape[3]
    num_grids = math.ceil(math.sqrt(num_filters))
    
    # -3 Create a figure with a grid
    fig, axes = plt.subplots(num_grids, num_grids)
    
    # -4 Plot the filter-weights
    for i, ax in enumerate(axes.flat):
        # Only plot valid weights
        if  i<num_filters:
            # get the weights for the i filter from the input channel
            # Check new_conv_layer()
            img = w[ : , : ,input_channel, i]
            
            #Plot
            ax.imshow(img, vmin = w_min, vmax = w_max, interpolation = 'nearest', cmap = 'seismic')
        # remove tick from plt
        ax.set_xticks([])
        ax.set_yticks([])
    
    plt.show()
```

### 2- Output of the conv layer

```python

def plot_conv_layer(layer, image):    
    # Assume layer is a TensorFlow op that outputs a 4-dim tensor
    # which is the output of a convolutional layer,
    # e.g. layer_conv1 or layer_conv2.

    # Create a feed-dict containing just one image.
    # Note that we don't need to feed y_true because it is
    # not used in this calculation.
    feed_dict = {x: [image]}

    # Calculate and retrieve the output values of the layer
    # when inputting that image.
    values = session.run(layer, feed_dict=feed_dict)

    # Number of filters used in the conv. layer.
    num_filters = values.shape[3]

    # Number of grids to plot.
    # Rounded-up, square-root of the number of filters.
    num_grids = math.ceil(math.sqrt(num_filters))
    
    # Create figure with a grid of sub-plots.
    fig, axes = plt.subplots(num_grids, num_grids)

    # Plot the output images of all the filters.
    for i, ax in enumerate(axes.flat):
        # Only plot the images for valid filters.
        if i<num_filters:
            # Get the output image of using the i'th filter.
            # See new_conv_layer() for details on the format
            # of this 4-dim tensor.
            img = values[0, :, :, i]

            # Plot image.
            ax.imshow(img, interpolation='nearest', cmap='binary')
        
        # Remove ticks from the plot.
        ax.set_xticks([])
        ax.set_yticks([])
    
    # Ensure the plot is shown correctly with multiple plots
    # in a single Notebook cell.
    plt.show()
```
